```python
name = input("Enter person name: ")

age = int(input("Enter person age: "))

height = float(input("Enter person height: "))

student = input("Enter whether you are student or not (Yes/No)")

print(name, age, height, student)



print(f"Name: {name} | Age: {age} | Height: {height} | Student: {student}")



age_in_months = age*12
age_in_days = age*365
age_div_7 = age%7
age_raise_pow_2 = age**2

print(f"Age in months: {age_in_months}\n"
f"Age in days: {age_in_days}\n"
f"The remainder when age is divided by 7: {age_div_7}\n"
f"Age raised to the power of 2: {age_raise_pow_2}")
```
